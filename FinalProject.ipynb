## Source Document Structure (1)

Provide a brief description of the internal structure of each document. That, describe the typical elements found in document and their relation to each other. For example, a corpus of letters might be described as having a date, an addressee, a salutation, a set of content paragraphs, and closing. If they are various structures, state that.

The constitution files have related but not perfectly identical internal structures. Each file begins with a document-level header that usually gives the country and year, such as `Afghanistan 2004`. Many constitutions then open with a preamble, which is an introductory statement of national principles, historical background, or political values. After that, the main body is often divided into larger organizational units such as chapters, parts, titles, or Roman-numeral sections. These larger divisions group related legal material, such as fundamental rights, executive power, legislative power, courts, amendment procedures, or transitional provisions.

Within those larger divisions, the main legal units are usually articles. Articles are the most common repeated structural unit in the corpus, but not every constitution labels them the same way. Some documents use `Section` headings, some include `Amendment` headings, and others use bare numbered headings such as `1. Name and territory`. The parser treats these article-like units as the main middle level of the OHCO hierarchy so that different national formats can be compared in one table.

Below the article or section level, the body text is organized into paragraph-like clauses. These clauses often contain the actual legal provisions. Some clauses are simple paragraphs, while others begin with numbered, lettered, or Roman-numeral subclauses such as `1.`, `a.`, or `IV.`. The parser preserves these as clause-level text rather than creating a separate subclause level. Finally, each clause is split into individual word tokens. For the project, these varied source structures are normalized into the OHCO hierarchy `Constitution -> Article -> Clause -> Token`. Larger divisions such as chapters, parts, titles, preambles, and amendments help the parser recognize structure, but they are not exported as required corpus columns because they are not used consistently across all constitutions.

The strength of this approach is that it creates a consistent table that can be compared across many countries even when their constitutions use different formatting conventions. It keeps the most important legal units, articles and clauses, available for analysis while avoiding a larger hierarchy that would be uneven across documents. The main tradeoff is simplification: chapter- or part-level thematic structure can matter for interpretation, but analyses based on the normalized OHCO will emphasize articles and clauses more than those larger divisions.

In addition to the constitution texts themselves, the project uses selected country-year variables from the V-Dem dataset at the document level. These variables, `v2x_regime`, `v2x_freexp_altinf`, and `v2x_rule`, are matched by `country` and `year_amended` and stored in the `LIB` table rather than the `CORPUS`. They are not part of the token sequence or OHCO structure, but they provide important political context for comparing constitutional language across different regime conditions.

- Description of TFIDIF formula ($\LaTeX$ OK): using `TfidfTransformer(norm=None, smooth_idf=True)` from SciKit-Learn, the table computes `tfidf(t, d) = tf(t, d) * idf(t)`, where `tf(t, d)` is the raw term count in document bag `d` and `idf(t) = log((1 + N) / (1 + df(t))) + 1`.
```python
from pathlib import Path
from collections import Counter


def report_raw_documents(folder_path):
    folder = Path(folder_path)
    files = [path for path in folder.iterdir() if path.is_file()]

    num_docs = len(files)
    total_bytes = sum(path.stat().st_size for path in files)
    total_mb = total_bytes / (1024 * 1024)
    formats = Counter(path.suffix.lower().lstrip(".") or "no extension" for path in files)

    print(f"Number of raw documents: {num_docs}")
    print(f"Total size of raw documents: {total_mb:.2f} MB")
    print("File format(s):")
    for file_format, count in sorted(formats.items()):
        label = "plaintext (.txt)" if file_format == "txt" else file_format
        print(f"- {label}: {count}")

    return {
        "num_docs": num_docs,
        "total_mb": total_mb,
        "formats": dict(formats),
    }


raw_report = report_raw_documents("individual_constitutions")
```
